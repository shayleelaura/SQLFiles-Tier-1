```python
# Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members! */

pd.read_sql_query('SELECT * sub.name, SUM( sub.revenue ) AS total_rev FROM (SELECT fac.name,CASE WHEN book.memid >0 THEN book.slots * membercost ELSE book.slots * guestcost END AS revenue FROM Bookings book JOIN Facilities fac ON book.facid = fac.facid) AS sub GROUP BY sub.nameHAVING SUM( sub.revenue ) <1000 ORDER BY total_rev', conn)
```
